Установка библиотек

```python
import sqlite3
import pandas as pd
from IPython.display import display
from ydata_profiling import ProfileReport
```

```python
import uuid
```

Функция для присвоения уникального id

```python
def generate_uuid():
    return str(uuid.uuid4())
```

Загрузка дата-сета для его обработки

```python
conn = sqlite3.connect('/Users/olesiaak/Desktop/ds/Проектная работа sql/Shark_tank.db', timeout=10)
cursor = conn.cursor()
```

```python
data_sharks = pd.read_csv('Shark Tank Australia dataset 2.csv',
                          sep=',' 
                         )
data_sharks = data_sharks.rename(columns=lambda x: x.lower().replace(' ', '_'))
```

```python
data_sharks.head(10)
```

```python
data_sharks = data_sharks.drop_duplicates()
```

```python
data_sharks.describe()
```

Генерируем уникальный id, удалям столбцы, которые нам не будут нужны

```python
data_sharks['id'] = [generate_uuid() for _ in range(len(data_sharks))]
data_sharks.drop(['episode_number',
                  'season_start',
                  'season_end',
                  'original_air_date',
                  'pitchers_state',
                  'multiple_entrepreneurs',
                  'royalty_deal',
                  'advisory_shares_equity',
                  'deal_has_conditions',
                  'steve_present',
                  'janine_present',
                  'naomi_present',
                  'glen_present',
                  'guest_present',
                  'andrew_present'], axis=1, inplace=True)
```

```python
list(data_sharks.columns.values)
```

Формируем дата-фрейм с общей информацией о проектах

```python
startups = data_sharks[['startup_name',
                        'industry',
                        'business_description',
                        'company_website',
                        'original_ask_amount',
                        'original_offered_equity',
                        'valuation_requested',
                        'viewers']].copy()
```

Меняем тип данных столбца - просмотры

```python
startups['viewers'] = startups['viewers'].str.replace(',', '').astype('float')
```

Генерируем уникальный id 

```python
startups['id'] = [generate_uuid() for _ in range(len(startups))]
```

Создаем дата-фрейм с данными об участниках. Присваиваем уникальный id, а также добавляем foreign_key startup_id

```python
pitchers = data_sharks[['pitchers_gender', 'pitchers_average_age', 'pitchers_city', 'entrepreneur_names']].copy()
pitchers['id'] = [generate_uuid() for _ in range(len(pitchers))]
pitchers['startup_id'] = startups['id']
pitchers
```

```python
startups['pitcher_id'] = pitchers['id']
```

Создаем дата-фрейм с данными об офферах. Присваиваем уникальный id, а также добавляем foreign_key во фрейм startups

```python
offers = data_sharks[[
    'received_offer',
    'accepted_offer',
    'total_deal_amount',
    'total_deal_equity',
    'deal_valuation',
    'number_of_sharks_in_deal',
    'investment_amount_per_shark',
    'equity_per_shark',
    'loan']].copy()
```

```python
offers['id'] = [generate_uuid() for _ in range(len(offers))]
```

```python
startups['offer_id'] = offers['id']
```

Создаем дата-фрейм с данными об инвестициях и Акулах. Присваиваем уникальный id, а также добавляем foreign_key startup_id

```python
investments = data_sharks[[
    'startup_name',
    'steve_investment_amount',
    'steve_investment_equity',
    'janine_investment_amount',
    'janine_investment_equity',
    'andrew_investment_amount',
    'andrew_investment_equity',
    'naomi_investment_amount',
    'naomi_investment_equity',
    'glen_investment_amount',
    'glen_investment_equity',
    'guest_investment_amount',
    'guest_investment_equity',
    'invested_guest_name']].copy()
```

```python
investments['id'] = [generate_uuid() for _ in range(len(investments))]
investments['startup_id'] = startups['id']
```

```python
investments.head(3)
```

```python
startups['investment_id'] = investments['id']
```

Создаем таблицы в базе данных для дальнейшей работы в Dbeaver

```python
startups.to_sql(con=conn, name='startups', index=False)
pitchers.to_sql(con=conn, name='pitchers', index=False)
offers.to_sql(con=conn, name='offers', index=False)
investments.to_sql(con=conn, name='investments', index=False)
conn.commit()
```
